```python
using System;

class Plant
{
    public string Name { get; set; }
    public string Type { get; set; }
    public float Height { get; set; }

    public Plant(string name, string type, float height)
    {
        Name = name;
        Type = type;
        Height = height;
    }

    public virtual void Bloom()
    {
        Console.WriteLine($"Растение {Name} цветет.");
    }

    public void Grow(float cm)
    {
        Height += cm;
        Console.WriteLine($"Растение {Name} выросло на {cm} см. Теперь высота: {Height} см.");
    }
}

class Flower : Plant
{
    public string Color { get; set; }
    public int PetalCount { get; set; }

    public Flower(string name, float height, string color, int petalCount)
        : base(name, "Цветок", height)
    {
        Color = color;
        PetalCount = petalCount;
    }

    public override void Bloom()
    {
        Console.WriteLine($"Цветок {Name} цветет! Цвет: {Color}, лепестков: {PetalCount}.");
    }

    public void Wither()
    {
        Console.WriteLine($"Цветок {Name} начинает увядать.");
    }
}

class Tree : Plant
{
    public int Age { get; set; }
    public string LeafType { get; set; }

    public Tree(string name, float height, int age, string leafType)
        : base(name, "Дерево", height)
    {
        Age = age;
        LeafType = leafType;
    }

    public override void Bloom()
    {
        Console.WriteLine($"Дерево {Name} покрывается новыми листьями! Тип листьев: {LeafType}.");
    }

    public void GrowOlder()
    {
        Age++;
        Console.WriteLine($"Дерево {Name} стало на год старше. Новый возраст: {Age} лет.");
    }
}

var rose = new Flower(name: "Роза", height: 30f, color: "Красный", petalCount: 20);
var pine = new Tree(name: "Сосна", height: 200f, age: 5, leafType: "Хвойные");

Console.WriteLine("=== Рост растений ===");
rose.Grow(5f);
pine.Grow(20f);

Console.WriteLine("\n=== Цветение (полиморфизм) ===");
Plant[] plants = { rose, pine };

foreach (Plant p in plants)
{
    p.Bloom();
}

Console.WriteLine("\n=== Специальные методы наследников ===");
rose.Wither();
pine.GrowOlder();
```
